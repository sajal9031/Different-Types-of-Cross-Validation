```python
df=pd.read_csv('cancer_dataset.csv')
df.head()
```

```python
X=df.iloc[:,2:]
X.head()
```

```python
y=df.iloc[:,1]
y.head()
```

```python
X.isnull().sum()
```

```python
X=X.dropna(axis=1)
```

```python
X.head()
```

```python
y.value_counts()
```

```python
### HoldOut Validation Approach- Train And Test Split
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=4)
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
result = model.score(X_test, y_test)
print(result)
```

```python
## K Fold Cross Validation
```

```python
from sklearn.model_selection import KFold
model=DecisionTreeClassifier()
kfold_validation=KFold(10)
```

```python
from sklearn.model_selection import cross_val_score
results=cross_val_score(model,X,y,cv=kfold_validation)
print(results)
print(np.mean(results))
```

```python
## Stratified K-fold Cross Validation
```

```python
from sklearn.model_selection import StratifiedKFold
skfold=StratifiedKFold(n_splits=5)
model=DecisionTreeClassifier()
scores=cross_val_score(model,X,y,cv=skfold)
print(np.mean(scores))
scores
```

```python
## Leave One Out Cross Validation(LOOCV)
```

```python
from sklearn.model_selection import LeaveOneOut
model=DecisionTreeClassifier()
leave_validation=LeaveOneOut()
results=cross_val_score(model,X,y,cv=leave_validation)
results
```

```python
print(np.mean(results))
```

```python
#Repeated Random Test-Train Splits
#(This technique is a hybrid of traditional train-test splitting and the k-fold cross-validation method.)
#(In this technique, we create random splits of the data in the training-test set manner and then repeat the process of splitting and evaluating the algorithm multiple times,
# (just like the cross-validation method.)
```

```python
from sklearn.model_selection import ShuffleSplit
model=DecisionTreeClassifier()
ssplit=ShuffleSplit(n_splits=10,test_size=0.30)
results=cross_val_score(model,X,y,cv=ssplit)
```
